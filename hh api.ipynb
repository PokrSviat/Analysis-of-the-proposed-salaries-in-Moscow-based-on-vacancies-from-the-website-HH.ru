```python
# Import the necessary libraries
import requests
import json
import time
import os
import pandas
import pandas as pd
from sqlalchemy import engine as sql
from IPython import display
import io
import numpy as np
```

```python
#Creating a list of metro names and IDs
metro_list = []
for j in range(0, len(metro[0]['lines'])):
    for z in range(0, len(metro[0]['lines'][j]['stations'])):
        #print(metro[0]['lines'][j]['stations'][z]['name'])
        metro_list.append(metro[0]['lines'][j]['stations'][z]['name'])
metro_list_id = []
for j in range(0, len(metro[0]['lines'])):
    for z in range(0, len(metro[0]['lines'][j]['stations'])):
        #print(metro[0]['lines'][j]['stations'][z]['name'])
        metro_list_id.append(metro[0]['lines'][j]['stations'][z]['id'])
```

```python
def getPage(metro, page = 0):
    """
    Creating a method to get a page with a list of vacancies.
    Arguments:
        page - The index of the page starts from 0. The default value is 0, the first page
        metro - A list of metro stations through which requests will be processed and vacancies will be uploaded for each metro station.
    """
    
    # Reference for GET request parameters
    params = {
        'text': '',                                # The filter text. we leave it empty to get all possible options
        'area': 1,                                 # The search is carried out by vacancies in the city of Moscow
        'page': page,                              # Index of the search page on HH
        'per_page': 100,                           # Number of vacancies on 1 page
        'metro': metro                             # Metro station
    }
    
    
    req = requests.get('https://api.hh.ru/vacancies', params) # Sending a request to the API
    data = req.content.decode() # We decode his answer so that the Cyrillic alphabet is displayed correctly
    req.close()
    return data
```

```python
# We start a cycle that incrementally requests pages based on requests.
for metro_id in metro_list_id:
    for page in range(0, 20):
        
        # Converting the text of the request response to the Python reference
        jsObj = json.loads(getPage(metro_id, page))
        
        # Save the files to the folder ./docs/pagination
        # We determine the number of files in the folder to save the document with the request response
        # The resulting value is used to form the name of the document
        nextFileName = './docs/pagination/{}.json'.format(len(os.listdir('./docs/pagination')))
        
        # We create a new document, write the request response into it, and then close it
        f = open(nextFileName, mode='w', encoding='utf8')
        f.write(json.dumps(jsObj, ensure_ascii=False))
        f.close()
        
        # Check to the last page if there are fewer than 2000 vacancies
        if (jsObj['pages'] - page) <= 1:
            break
        
        # An optional delay, but in order not to load the hh services, we will leave it.
        time.sleep(1)
    
print('Старницы поиска собраны')
```

```python
# We get a list of previously created files with a list of vacancies and go through it in a loop
for fl in os.listdir('./docs/pagination'):
    
    # Open the file, read its contents, close the file
    f = open('./docs/pagination/{}'.format(fl), encoding='utf8')
    jsonText = f.read()
    f.close()
    
    # Converting the received text into a directory object
    jsonObj = json.loads(jsonText)
    
    # We receive and go through the list of vacancies directly
    for v in jsonObj['items']:
        
        # We turn to the API and get detailed information on a specific vacancy
        req = requests.get(v['url'])
        data = req.content.decode()
        req.close()
        
        # Creating a file in json format with the job ID as the name
        # We write the request response to it and close the file
        fileName = './docs/vacancies/{}.json'.format(v['id'])
        f = open(fileName, mode='w', encoding='utf8')
        f.write(data)
        f.close()
        
        time.sleep(0.25)
        
print('Вакансии собраны')
```

```python
# We will display the progress in the output
# To do this, we will find out the total number of files that need to be processed
# We will set the counter of processed files to zero
cnt_docs = len(os.listdir(r'./docs/vacancies'))
i = 0
 
# We go through all the files in the vacances folder
for fl in x:
     
    # Open, read and close the file
    with open(r'./docs/vacancies/{}'.format(fl), encoding='utf-8') as f:
        jsonText = f.read()
        f.close()
     
    # The text of the file is translated into the directory
    jsonObj = json.loads(jsonText)
    
    if 'errors' in jsonObj:
        continue
    else:
        # Filling in the lists for the tables
        IDs.append(jsonObj['id'])
        names.append(jsonObj['name'])
        descriptions.append(jsonObj['description'])
        employment.append(jsonObj['employment'])
        schedule.append(jsonObj['schedule'])
        professional_roles.append(jsonObj['professional_roles'])
        accept_handicapped.append(jsonObj['accept_handicapped'])
        accept_kids.append(jsonObj['accept_kids'])
        experience.append(jsonObj['experience'])
        address.append(jsonObj['address'])
        department.append(jsonObj['department'])
        salary.append(jsonObj['salary'])
        published_at.append(jsonObj['published_at'])
        typev.append(jsonObj['type'])
        has_test.append(jsonObj['has_test'])
        specializations.append(jsonObj['specializations'])
        working_days.append(jsonObj['working_days'])
        working_time_intervals.append(jsonObj['working_time_intervals'])
        working_time_modes.append(jsonObj['working_time_modes'])
        accept_temporary.append(jsonObj['accept_temporary'])
        driver_license_types.append(jsonObj['driver_license_types'])
        languages.append(jsonObj['languages'])
        
        # Since skills are stored as an array, we go through it in a loop
        for skl in jsonObj['key_skills']:
            skills_vac.append(jsonObj['id'])
            skills_name.append(skl['name'])
         
        # We increase the counter of processed files by 1, clear the cell output and display the progress
        i += 1
        #print(fl)
        display.clear_output(wait=True)
        display.display('Готово {} из {}'.format(i, cnt_docs)
 
# We create a dataframe, which we then save to the database in the vacances table
df = pd.DataFrame({'id': IDs, 'name': names, 'description': descriptions, 'employment': employment, 'schedule': schedule,
                   'accept_handicapped': accept_handicapped, 'accept_kids': accept_kids, 'experience': experience,'salary': salary,
                   'published_at': published_at,'type': typev, 'has_test': has_test, 'specializations': specializations,
                   'working_days': working_days,' working_time_intervals':  working_time_intervals, 'professional_roles':professional_roles,
                   'working_time_modes': working_time_modes,'accept_temporary': accept_temporary, 'driver_license_types':driver_license_types,
                   'languages':languages})
df1 = pd.DataFrame({'vacancy': skills_vac, 'skill': skills_name})

# We display a message about the end of the program
display.clear_output(wait=True)
display.display('Вакансии загружены в БД')
```
